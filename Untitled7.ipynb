```python
import pandas as pd
import numpy as np

# Load datasets
sales = pd.read_csv("sales.csv")
products = pd.read_csv("product.csv")
customers = pd.read_csv("customers.csv")

# Check data
print(sales.head())
print(sales.info())

# Remove duplicates
sales.drop_duplicates(inplace=True)

# Handle missing values
sales.fillna(0, inplace=True)

# Convert date column
sales['sale_date'] = pd.to_datetime(sales['sale_date'])

# Merge products to get unit price
sales = pd.merge(sales, products[['product_id', 'price']], on='product_id', how='left')

# Create Total Sale Amount
sales['total_sale_amount'] = (
    sales['quantity_sold'] * sales['price']
)

# Extract Month and Year
sales['month'] = sales['sale_date'].dt.month
sales['year'] = sales['sale_date'].dt.year

# Save cleaned file
sales.to_excel("cleaned_sales.xlsx", index=False)

print("Data cleaned successfully")
```
